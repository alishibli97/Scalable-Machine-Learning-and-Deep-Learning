```python
df = pd.read_csv("/content/drive/MyDrive/Scalable ML/project/data/netflix_titles.csv")
```

```python
df = df.dropna()
df.reset_index(inplace=True)
```

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
data = []
for i,descr in enumerate(df.description):
  print(f"Finished {i}/{len(df)}")
  emb = model.encode(descr)
  data.append(emb)
import numpy as np
data = np.array(data)
print(data.shape)
```

```python
np.save("embeddings", data)
```

```python
embeddings = np.load("embeddings.npy")
embeddings.shape
```
